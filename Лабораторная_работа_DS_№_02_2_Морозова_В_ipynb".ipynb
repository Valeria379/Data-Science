## 2.2.1: Reading and Writing Data

Файл [grades.csv](grades.csv) представляет собой файл с именами учащихся и буквенными оценками:

```
Alice; A
Bob; B
Robert; A
Richard; C
```

Записать данные из файла в массив. Добавьте средний балл учащегося в строку учащегося (A=4,B=3,C=2,D=1).

Подсказка: функция [strip()](https://docs.python.org/3/library/stdtypes.html#str.strip) удаляет конечные пробелы из строки.

Запишите этот файл в новый файл `grades_gpa.csv`.

````python
gpas = {"A":4, "B":3, "C":2, "D":1}
````

Загрузим файл

````python
from google.colab import files
uploated = files.upload()
````

Запишем данные из файла в массив

````python
# импортировать библиотеку csv
import csv

# инициализировать массив верхнего уровня
array = []

with open("grades.csv") as csvfile:
    filereader = csv.reader(csvfile, delimiter=';')

    for row in filereader:
      row.append(gpas[row[1].strip()])
      array.append(row)
````

````python
array
````

Запишем этот файл в новый файл

````python
with open('my_file.csv', 'w') as new_file:
    new_file.write("[['Alice', ' A', 4], ['Bob', ' B', 3], 'Robert', ' A', 4], ['Richard', ' C', 2]]")
````


 * Создайте новый фрейм данных на основе набора данных hit_albums, который содержит только исполнителей, у которых в списке есть как минимум два альбома.

Посчитаем количество альбомов у каждого исполнителя

````python
count_albums = hit_albums['Artist'].value_counts()
count_albums
````

````python
filter_artist = count_albums[count_albums >= 2].reset_index()
filter_artist
````

Переименуем столбцы и выведем итоговый датафрейм, который содержит только исполнителей, у которых в списке есть как минимум два альбома.

````python
filter_artist.rename(columns={'index':'Artist', 'Artist':'Count_albom'}, inplace=True)
filter_artist
````
